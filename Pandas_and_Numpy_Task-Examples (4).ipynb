## 2. Creating arrays (array , arange, linspace, zeros, ones, random.randint ) 

```python
# 1. array() : Creates a array fro a list or tuple. 
import numpy as np
a = np.array([1,2,3,4,5])
print(a)

# 2. arange(): Creates an array with values in a given range.
b = np.arange(1,11)
print(b)

# 3. linspace():Creates equally spaced values between two numbers.
c = np.linspace(2,20,5)
print(c)

# 4. zeros():  Creates an array filled with zeros.
d = np.zeros(10)
print(d)

# 5. ones():   Creates an array filled with ones.
e = np.ones(5)
print(e)

# 6. random.randint(): Creates an array of random integers. 
a = np.random.randint(1, 10, 5)
print(a)
```

### 3. Array attributes ( ndim, shape, size, dtype, itemsize )

## 5. Slicing - Multiple values

```python
import numpy as np

arr = np.array(['Banana', 'Apple', 'Mango', 'Orange'])

# index 1 to 2 (stop not included)
print(arr[1:3])

# start to index 1
print(arr[:2])

# index 2 to end
print(arr[2:])

# every second element (step = 2)
print(arr[::2])

# negative slicing (from end side)
print(arr[-3:-1])
```

## 7. Boolean (Mask) Indexing / filtering

## 10. Copy vs Reference in NumPy (.copy)

```python
# 1. Reference (No .copy())
import numpy as np

arr1 = np.array([10, 20, 30, 40])

# creating reference (same memory location)
arr2 = arr1

# modifying arr2
arr2[0] = 99

print("arr1:", arr1)
print("arr2:", arr2)

# True Copy (.copy())
arr1 = np.array([10, 20, 30, 40])

# creating an independent copy
arr2 = arr1.copy()

# modifying arr2
arr2[0] = 99

print("arr1:", arr1)
print("arr2:", arr2)
```

## 14. Reshaping, Flatten vs Ravel

```python
# Reshaping (reshape): Changes the shape of the array (data size remains same)
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

# reshaping into 2 rows and 3 columns
new_arr = arr.reshape(2, 3)

print(new_arr)

# Flatten: Converts array into 1D array (copy)
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

flat = arr.flatten()

print(flat)

# Ravel: Converts array into 1D array (view)
rav = arr.ravel()

print(rav)
```

# Section 4: Combining DataFrames

```python
temp_df = pd.DataFrame({
    'city': ['Pune', 'Nagpur', 'Nashik', 'Kolhapur'],
    'temp': [30, 38, 29, 31]
})

hum_df = pd.DataFrame({
    'city': ['Pune', 'Nagpur', 'Aurangabad', 'Solapur'],
    'humidity': [72, 58, 65, 60]
})

print(temp_df)
print()
print(hum_df)
```

```python
# concat: stacks rows on top of each other (note duplicate index unless ignore_index=True)
pd.concat([temp_df, hum_df],
ignore_index=True)
```

```python
# concat along columns (axis=1) - aligns by index position
pd.concat([temp_df, hum_df], axis=1)
```

```python
# merge: outer join keeps ALL keys from both sides, filling NaN where unmatched
pd.merge(temp_df, hum_df, on='city', how='outer')
```

```python
# merge: inner join keeps ONLY keys present in both DataFrames 
pd.merge(temp_df, hum_df, on='city', how='inner')
```

```python
# merge: left / right joins
print(pd.merge(temp_df, hum_df, on='city', how='left')) 
print()
print(pd.merge(temp_df, hum_df, on='city', how='right'))
```

```python
# join: merges on the INDEX instead of a column
left = temp_df.set_index('city') 
right = hum_df.set_index('city') 
left.join(right, how='outer')
```

```python
import numpy as np

# User-item interactions (user_id, item_id)
interactions = np.array([
    [0, 1],
    [0, 2],
    [1, 1],
    [1, 3],
    [2, 2],
    [2, 3]
])

# Number of items
num_items = 4

# Create an empty co-occurrence matrix
co_matrix = np.zeros((num_items, num_items), dtype=int)

# Build co-occurrence matrix
for user in np.unique(interactions[:, 0]):

    # Get all items purchased by the current user
    items = interactions[interactions[:, 0] == user][:, 1]

    # Increase the count for every item pair
    np.add.at(co_matrix, (items[:, None], items), 1)

print("Co-occurrence Matrix:")
print(co_matrix)
```

## 6. Computing IoU (Intersection over Union) for bounding boxes in object detection — pure vectorized array math.

```python
import numpy as np

# Bounding boxes in the format: [x1, y1, x2, y2]
box1 = np.array([
    [10, 10, 50, 50],
    [20, 20, 60, 60]
])

box2 = np.array([
    [30, 30, 70, 70],
    [25, 25, 55, 55]
])

# Calculate intersection coordinates
x1 = np.maximum(box1[:, 0], box2[:, 0])
y1 = np.maximum(box1[:, 1], box2[:, 1])
x2 = np.minimum(box1[:, 2], box2[:, 2])
y2 = np.minimum(box1[:, 3], box2[:, 3])

# Calculate intersection area
intersection = np.maximum(0, x2 - x1) * np.maximum(0, y2 - y1)

# Calculate individual box areas
area1 = (box1[:, 2] - box1[:, 0]) * (box1[:, 3] - box1[:, 1])
area2 = (box2[:, 2] - box2[:, 0]) * (box2[:, 3] - box2[:, 1])

# Calculate union area
union = area1 + area2 - intersection

# Compute IoU
iou = intersection / union

print("IoU Values:")
print(iou)
```

## 11. Computing per-class precision/recall directly from a confusion matrix using row/column sums.

```python
import numpy as np

# Confusion Matrix
# Rows = Actual Class
# Columns = Predicted Class
conf_matrix = np.array([
    [50, 2, 3],
    [4, 45, 1],
    [2, 3, 40]
])

# True Positives (Diagonal elements)
tp = np.diag(conf_matrix)

# Predicted Positives (Column sums)
predicted = np.sum(conf_matrix, axis=0)

# Actual Positives (Row sums)
actual = np.sum(conf_matrix, axis=1)

# Compute Precision and Recall
precision = tp / predicted
recall = tp / actual

print("Confusion Matrix:\n", conf_matrix)
print("\nPrecision per Class:")
print(precision)

print("\nRecall per Class:")
print(recall)
```

## 13	Batch-processing embeddings — computing cosine similarity between a query embedding and a matrix of stored embeddings in one vectorized operation (used in semantic search/recommendation).

```python
import numpy as np

# Query embedding (1 vector)
query = np.array([0.2, 0.4, 0.6])

# Stored embeddings (4 vectors)
embeddings = np.array([
    [0.1, 0.3, 0.5],
    [0.2, 0.5, 0.7],
    [0.9, 0.1, 0.2],
    [0.3, 0.4, 0.6]
])

# Compute dot product
dot_product = embeddings @ query

# Compute vector norms
embedding_norms = np.linalg.norm(embeddings, axis=1)
query_norm = np.linalg.norm(query)

# Compute cosine similarity
cosine_similarity = dot_product / (embedding_norms * query_norm)

print("Cosine Similarity:")
print(cosine_similarity)
```

## 20.	Computing dataset statistics for a data report — mean , std , min , max , percentiles per column, exactly the kind of summary you'd put in an EDA section of a project report.

```python
import numpy as np

# Sample dataset
data = np.array([
    [10, 20, 30],
    [15, 25, 35],
    [20, 30, 40],
    [25, 35, 45],
    [30, 40, 50]
])

# Compute statistics column-wise
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
minimum = np.min(data, axis=0)
maximum = np.max(data, axis=0)
percentiles = np.percentile(data, [25, 50, 75], axis=0)

print("Mean:", mean)
print("Standard Deviation:", std)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("\n25th Percentile:", percentiles[0])
print("50th Percentile (Median):", percentiles[1])
print("75th Percentile:", percentiles[2])
```
