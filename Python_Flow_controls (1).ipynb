```python
            condition true aasel tr true execute honar , jar condition false aasel tr else execute honar.
```

### a. for loop

```python
     Iterative Statement mhanje ekhada code block punha punha (repeat) execute karnyasathi vaparla jato.
```

```python
                for variable in range(start, stop):
                        statement
```

```python
for i in range(1,10):
    print(i)
```

```python
for a in range(20):
    if a % 2 != 0:
        print(a)
```

# Quick Reference

##### 1. Deference Between for and while loop.?

| For Loop | While Loop |
|----------|------------|
| Commonly used with `range()`. | It is based on a condition. |
| Runs a fixed number of times. | Runs until a condition becomes False. |
| For loop tharavik vela execute hoto. | While loop condition True asel toparyant execute hoto. |
| Example: `for a in range(10): print(a)` | Example: `a = 1` <br> `while a <= 50:` <br> `print(a)` <br> `a += 1` |
