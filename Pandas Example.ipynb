# Section 4: Combining DataFrames

```python
temp_df = pd.DataFrame({
    'city': ['Pune', 'Nagpur', 'Nashik', 'Kolhapur'],
    'temp': [30, 38, 29, 31]
})

hum_df = pd.DataFrame({
    'city': ['Pune', 'Nagpur', 'Aurangabad', 'Solapur'],
    'humidity': [72, 58, 65, 60]
})

print(temp_df)
print()
print(hum_df)
```

```python
# concat: stacks rows on top of each other (note duplicate index unless ignore_index=True)
pd.concat([temp_df, hum_df],
ignore_index=True)
```

```python
# concat along columns (axis=1) - aligns by index position
pd.concat([temp_df, hum_df], axis=1)
```

```python
# merge: outer join keeps ALL keys from both sides, filling NaN where unmatched
pd.merge(temp_df, hum_df, on='city', how='outer')
```

```python
# merge: inner join keeps ONLY keys present in both DataFrames 
pd.merge(temp_df, hum_df, on='city', how='inner')
```

```python
# merge: left / right joins
print(pd.merge(temp_df, hum_df, on='city', how='left')) 
print()
print(pd.merge(temp_df, hum_df, on='city', how='right'))
```

```python
# join: merges on the INDEX instead of a column
left = temp_df.set_index('city') 
right = hum_df.set_index('city') 
left.join(right, how='outer')
```

# Section 5: Importing / Exporting Data

```python
# --- Build a synthetic student dataset ---

np.random.seed(42)
n = 200

df = pd.DataFrame({
    'Student_ID': np.arange(1, n + 1),
    'Age': np.random.randint(17, 25, n),
    'Gender': np.random.choice(['Male', 'Female'], n, p=[0.55, 0.45]),
    'Maths': np.random.randint(35, 101, n),
    'Science': np.random.randint(35, 101, n),
    'English': np.random.randint(35, 101, n),
    'Attendance': np.round(np.random.uniform(60, 100, n), 1),
    'Study_Hours': np.round(np.random.uniform(1, 8, n), 1),
    'Sports': np.random.choice(['Yes', 'No'], n, p=[0.4, 0.6]),
    'Result': np.random.choice(['Pass', 'Fail'], n, p=[0.85, 0.15])
})

# Inject a few missing values and duplicate rows
df.loc[[5, 18, 50, 95], 'Study_Hours'] = np.nan
df = pd.concat([df, df.iloc[[10, 10, 25]]], ignore_index=True)

print(df.shape)
print(df.head())
```

```python
# Export to CSV / Excel / JSON

df.to_csv('student_data.csv', index=False)
df.to_excel('student_data.xlsx', index=False)
df.to_json('student_data.json', orient='records')

print("Files written: student_data.csv / .xlsx / .json")

# Read them back

df_csv = pd.read_csv('student_data.csv')
df_xlsx = pd.read_excel('student_data.xlsx')
df_json = pd.read_json('student_data.json')

print(df_csv.shape, df_xlsx.shape, df_json.shape)

df_csv.head(3)
```

# Section 9: map, apply, DataFrame.map, and Lambda Functions

```python
# map() on a Series with a dictionary 
gender_code = {'Male': 2, 'Female' : 1}
df['Gender_code'] = df['Gender'].map(gender_code) 
df[['Gender', 'Gender_code']].head()
```

```python
# map() with a function / lambda
df['Age_in_months'] = df['Age'].map(lambda x: x * 14) 
df[['Age', 'Age_in_months']].head()
```

```python
# apply() on a Series - equivalent to map() for simple functions 
df['M1'] = df['Maths'].apply(lambda x: x ** 4)
df[['Maths', 'M1']].head()
```

```python
# apply() on a DataFrame, row-wise (axis=1)

def student_risk(row):
    if row['Maths'] < 40 or row['Attendance'] < 75:
        return 'High'
    else:
        return 'Low'

df['Risk'] = df.apply(student_risk, axis=1)

df[['Maths', 'Attendance', 'Risk']].head()
```

```python
# apply() on a DataFrame, column-wise (axis=0, the default) - function receives a full column
df[['Age', 'Maths', 'Science']].apply(lambda col: col.max() - col.min())
```

```python
# DataFrame.map() - element-wise transform across the WHOLE frame (numeric columns shown here)
df[['Age', 'Maths', 'Science']].map(lambda x: round(x, 1))
```

```python
# Cleanup helper columns
df.drop(columns=['Gender_code', 'Age_in_months', 'Risk'], inplace=True, errors='ignore')

print(df.head())
```

# Section 14: GroupBy — Split, Apply, Combine

```python
# Split + apply (mean) + combine, in one line 
df.groupby('Gender')['Age'].mean()
```

```python
	# size() - count of rows per group (includes NaN counts unlike count())
df.groupby('Gender').size()
```

```python
	# get_group() - pull out a single group as its own DataFrame
df.groupby('Gender').get_group('Female').head()
```

```python
		# agg() - apply multiple aggregation functions at once 
df.groupby('Gender')['Age'].agg(['mean', 'min', 'max', 'count'])
```

```python
# agg() with different functions per column df.groupby('Gender').agg({
summary = df.groupby('Gender').agg({
    'Age': 'mean',
    'Maths': ['mean', 'max'],
    'Result': lambda x: (x == 'Pass').sum()
})

print(summary)
```

```python
# transform() - returns an output the SAME SHAPE as the input (broadcast back to every row)
df['Age']	
df.groupby('Gender')['Age'].transform('mean') 
df[['Gender', 'Age', 'Age']].head()
```

```python
# filter() - keep only groups that satisfy a condition 
df.groupby('Gender').filter(lambda g: g['Age'].mean() > 4).head()
```
