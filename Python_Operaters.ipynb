## Relational Operator

```python
        Relational operatpor is used to compare the two values.
```

```python
1. < --> Greater than
2. <== --> Greater than and eqal to
3. > --> Less Than
4. >= --> Less than and equal to
5. == --> doubale equal to
6. != --> not equal to
```

```python
# Relational operator with numbers
a = 50
b = 20

print("a<b:",a<b)    # false
print("a<=b:",a<b)   # fales
print("a>b:",a>b)    # True
print("a>=b:",a>=b)  # true
print("a==b:",a==b)  # false 
print("a!=b:",a!=b)  # true
```

```python
# Relational operator with string
a = "Prajakta"
b = "Srushti"

print("a<b:",a<b)    # True
print("a<=b:",a<b)   # True
print("a>b:",a>b)    # false
print("a>=b:",a>=b)  # false
print("a==b:",a==b)  # false 
print("a!=b:",a!=b)  # true
```

## 4 . Bitwise Operator

```python
         Bitwise Operations means perform the operations in binary number.
```

```python
1. & --> Bitwise AND = Both bits 1 aastil tar result 1
2. | --> Bitwise OR = at least one bit 1 aasel tar result 1 yeto
3. ^ --> Bitwise XOR = bits different aastil tar result 1 yeto
4. ~ --> Bitwise Complement = bits reverse karto
5. << --> Left Shift = bits left side la shift karto
5. >> --> Right Shift = bits right side la shift karto
```

```python
x = 25
y = 50

print("x&y:",x&y)
print("x|y:",x|y)
print("x^y:",x^y)
print("~y:",~y)
print("x<<y:",x<<y)
print("x>>y:",x>>y)
```

## Assignment Operator

```python
        Assignment Operator is used to assign the value of variable.
```

```python
= --> Assign value 
+= --> Add and Assign
-= --> substract and assign
/= --> divude and assign
//= --> floor and asign
%= --> module and assign
&= --> AND and assign
^= --> OR and assign
|= --> NOT and assign
* --> Multipy and assign
```

```python
a = 20
b = 50

a += b      # a = 70
print("a+=b:", a)

a = b       # a = 50
print("a=b:", a)

a += b      # a = 100
print("a-=b:", a)   # (इथे तू += वापरलं आहे, -= नाही)

a /= b      # a = 100/50 = 2.0
print("a/=b:", a)

a //= b     # a = 2.0//50 = 0.0
print("a//=b:", a)

a %= b      # a = 0.0%50 = 0.0
print("a%=b:", a)
```
