# Seaborn Visualization using 5 Different Datasets

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set_theme(style="whitegrid")
```

# Dataset 1: Anscombe Dataset

## 1. Plot: Line Plot

```python
anscombe = sns.load_dataset("anscombe")

plt.figure(figsize=(8,5))

sns.lineplot(
    data=anscombe[anscombe["dataset"]=="I"],
    x="x",
    y="y"
)

plt.title("Line Plot - Anscombe Dataset")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

## 2. Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
anscombe = sns.load_dataset("anscombe")

# Pair Plot
sns.pairplot(
    anscombe,
    hue="dataset"
)

plt.show()
```

##  3. Histogram plot

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Select Dataset I
data = anscombe[anscombe["dataset"] == "I"]

# Create Histogram
plt.figure(figsize=(8,5))

sns.histplot(
    data=data,
    x="y",
    bins=8,
    kde=True,
    color="skyblue"
)

# Add title and labels
plt.title("Histogram of Y Values - Anscombe Dataset")
plt.xlabel("Y Values")
plt.ylabel("Frequency")

# Display plot
plt.show()
```

## 4. Box Plot

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Select Dataset I
data = anscombe[anscombe["dataset"] == "I"]

# Create Box Plot
plt.figure(figsize=(7,5))

sns.boxplot(
    data=data,
    y="y",
    color="lightgreen"
)

# Add title and labels
plt.title("Box Plot of Y Values - Anscombe Dataset")
plt.ylabel("Y Values")

# Display plot
plt.show()
```

## 5. Relplot

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Create Relplot
sns.relplot(
    data=anscombe,
    x="x",
    y="y",
    col="dataset",
    kind="scatter",
    height=4,
    aspect=1
)

# Add title
plt.suptitle("Relplot of Anscombe Dataset", y=1.05)

# Display plot
plt.show()
```

## 6. Join Plot

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Select Dataset I
data = anscombe[anscombe["dataset"] == "I"]

# Create Joint Plot
sns.jointplot(
    data=data,
    x="x",
    y="y",
    kind="scatter",
    height=6,
    color="blue"
)

# Display plot
plt.show()
```

## 7. Violin Plot 

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Create Violin Plot
plt.figure(figsize=(8,5))

sns.violinplot(
    data=anscombe,
    x="dataset",
    y="y",
    palette="Set2"
)

# Add title and labels
plt.title("Violin Plot of Y Values - Anscombe Dataset")
plt.xlabel("Dataset")
plt.ylabel("Y Values")

# Display plot
plt.show()
```
