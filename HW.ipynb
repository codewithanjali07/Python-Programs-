| Shallow Copy | Deep Copy |
|--------------|-----------|
| Creates a new object. | Creates a new object. |
| Nested objects are shared. | Nested objects are copied separately. |
| Changes in nested objects affect both copies. | Changes in nested objects do not affect the original object. |
| Uses `copy.copy()` | Uses `copy.deepcopy()` |
| Faster and uses less memory. | Slower and uses more memory. |
| Copies only the outer object. | Copies the complete object including nested objects. |

```python
import copy

a = [[10, 20], [30, 40]]

# Shallow Copy
b = copy.copy(a)

# Deep Copy
c = copy.deepcopy(a)

b[0][0] = 99
c[1][1] = 88

print("Original:", a)
print("Shallow:", b)
print("Deep:", c)
```
