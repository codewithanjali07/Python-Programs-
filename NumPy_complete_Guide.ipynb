## 7. The ndarray and why NumPy is memory-efficint ?

```python
ndarray (N-dimensional Array) is the main data structure in NumPy.
It stores elements of the same data type in a single array and supports 1D, 2D, and multi-dimensional arrays.
```

#### Example :

```python
import numpy as np

arr = np.array([10,20,30,40,50])
print(arr)
```

### Why is NumPy Memory-Efficient?

##### NumPy is memory-efficient because:

- It stores data in continuous memory locations.
- It stores only the same data type, reducing memory usage.
- It uses less memory than Python lists.
- It avoids storing extra information for each element.
- It provides fast access to data because of its memory layout.
