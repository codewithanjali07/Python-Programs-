# Seaborn Visualization using 5 Different Datasets

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set_theme(style="whitegrid")
```

# Dataset 1: Anscombe Dataset
# Plot: Line Plot

```python
anscombe = sns.load_dataset("anscombe")

plt.figure(figsize=(8,5))

sns.lineplot(
    data=anscombe[anscombe["dataset"]=="I"],
    x="x",
    y="y"
)

plt.title("Line Plot - Anscombe Dataset")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Dataset 2: Attention Dataset
# Plot: Bar Plot

```python
attention = sns.load_dataset("attention")

plt.figure(figsize=(8,5))

sns.barplot(
    data=attention,
    x="solutions",
    y="score",
    hue="attention"
)

plt.title("Bar Plot - Attention Score")
plt.xlabel("Solutions")
plt.ylabel("Score")
plt.show()
```

# Dataset 3: Car Crashes Dataset
# Plot: Scatter Plot

```python
car_crashes = sns.load_dataset("car_crashes")

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=car_crashes,
    x="speeding",
    y="total",
    size="alcohol"
)

plt.title("Scatter Plot - Speeding vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
plt.show()
```

# Dataset 4: Planets Dataset
# Plot: Histogram

```python
planets = sns.load_dataset("planets")

plt.figure(figsize=(8,5))

sns.histplot(
    data=planets,
    x="year",
    bins=20
)

plt.title("Histogram - Planet Discovery Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

# Dataset 5: Dow Jones Dataset
# Plot: Box Plot

```python
dowjones = sns.load_dataset("dowjones")

plt.figure(figsize=(8,5))

sns.boxplot(
    data=dowjones,
    y="Price"
)

plt.title("Box Plot - Dow Jones Price")
plt.ylabel("Price")
plt.show()
```
