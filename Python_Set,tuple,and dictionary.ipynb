### Special Care: Single-Valued Tuple

```python
# without trailing Comma - NOT a tuple

t = (20,)
print(t)
print(type(t))
print(id(t))
```

```python
# with trailing comma - IS a tuple

a = (10,20,30,)
print(a)
print(type(a))
print(id(a))
```

### Which are the following are valid tuples ?

| Tuple | Valid / Invalid | Reason |
|-------|-----------------|--------|
| `()` | ✅ Valid | Empty tuple |
| `(10, 20, 30)` | ✅ Valid | Multiple elements |
| `10, 20, 30` | ✅ Valid | Parentheses are optional |
| `(10,)` | ✅ Valid | Single-element tuple (comma is required) |
| `(10)` | ❌ Invalid as a tuple | Treated as an integer, not a tuple |
| `("Apple", "Mango")` | ✅ Valid | String tuple |
| `(1, "Hello", 3.5, True)` | ✅ Valid | Mixed data types are allowed |
| `([1, 2], (3, 4))` | ✅ Valid | Tuple can contain a list and another tuple |

# List vs Tuple

| Feature | List | Tuple |
|---------|------|-------|
| **Syntax** | `[]` | `()` |
| **Mutable** | ✅ Yes | ❌ No |
| **Add Elements** | `append()`, `extend()` | Not Allowed |
| **Remove Elements** | `remove()`, `pop()` | Not Allowed |
| **Performance** | Slower | Faster |
| **Memory Usage** | More Memory | Less Memory |
| **Methods** | Many Methods | `count()`, `index()` |
| **Use Case** | Changeable Data | Fixed Data |

# Part 2: Set Data Structure

```python
s = {10,20,30,40,50}
print(s)
print(type(s))
print(id(s))
```

```python
# from a list (removes duplicats automatically)
l = [1,2,3,4,1,2,5,6]
s = set(l)
print(s)

# from ranges 
s = set(range(0,10,2))
print(s)
```

```python
# creates a DICT , not a set
s = {}
print(s)
print(type(s))

# use set for empty set()
s = set()
print(s)
print(type(s))
```

## Important Functions of set

```python
# add() - add the element
s = {10,20,30,40}
s.add(50)
print(s)
```

```python
# update() - The update() method is used to add multiple elements to a set.
s = {10,20,30,40,50}
l = [1,2,3,4,5]
s.update(l,range(5))
print(s)
```

```python
# add() vs update()
# add() vs update()

fruits = {"Apple", "Mango"}

# add() → Adds one element
fruits.add("Banana")
print("After add():", fruits)

# update() → Adds multiple elements
fruits.update(["Orange", "Grapes", "Pineapple"])
print("After update():", fruits)
```

```python
# copy() - The copy() method is used to create a duplicate (copy) of a set.
s ={10,20,30}
s1 = s.copy()
print(s)
print(s1)
```

```python
# pop() - reomoves and returns are randam element
s = {10,20,30,40,50,60,70}
print(s)
print(s.pop())
print(s)
```

```python
# remove() - Remove for specified element
s = {11,22,33,44,55,66}
print(s)
s.remove(11)
print(s)
```

# remove() vs pop() vs discard()

| Feature                   | `remove()`               | `pop()`                                   | `discard()`              |
| ------------------------- | ------------------------ | ----------------------------------------- | ------------------------ |
| Used With                 | List                     | List                                      | Set                      |
| Purpose                   | Removes a specific value | Removes an element by index               | Removes a specific value |
| Syntax                    | `list.remove(value)`     | `list.pop(index)`                         | `set.discard(value)`     |
| Argument                  | Value                    | Index (optional)                          | Value                    |
| Default Behavior          | No default               | Removes last element if no index is given | No default               |
| Returns Value?            | ❌ No                     | ✅ Yes (returns removed element)           | ❌ No                     |
| Error if Item Not Found?  | ✅ Yes (`ValueError`)     | ✅ Yes (`IndexError` for invalid index)    | ❌ No error               |
| Modifies Original Object? | ✅ Yes                    | ✅ Yes                                     | ✅ Yes                    |


```python
# clear() - cler the set elements
a = { 10,20,30,40,50 }
print(a)
a.clear()
print(a)
```

## Mathamatical Operations in Set

```python
A = {10, 20, 30, 40}
B = {30, 40, 50, 60}

# union() - Combines all unique elements from both sets
print("Union:", A | B)

# insersection() - Returns common elements
print("Intersection:", A & B)

# Difference() -Returns elements in the first set but not in the second
print("Difference (A-B):", A - B)
print("Difference (B-A):", B - A)

# Symmetric Difference() - Returns elements that are in either set, but not in both
print("Symmetric Difference:", A ^ B)
```

# Progaram: Student name and marks

```python
student = {}

name = input("Enter student name: ")
marks = int(input("Enter student marks: "))

student["Name"] = name
student["Marks"] = marks

print("\nStudent Details")
print(student)
```

```python
# Store Details of Multiple Students
students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter student marks: "))
    students[name] = marks

print("\nStudent Records")
for name, marks in students.items():
    print(name, ":", marks)
```

# Tuple vs Dictionary vs Set

| Feature                 | Tuple                | Set                                         | Dictionary                                           |      |
| ----------------------- | -------------------- | ------------------------------------------- | ---------------------------------------------------- | ---- |
| **Syntax**                  | `( )`                | `{ }`                                       | `{key: value}`                                       |      |
| **Example**                 | `(10, 20, 30)`       | `{10, 20, 30}`                              | `{"Name":"Anjali", "Age":22}`                        |      |
| **Data Storage**            | Elements             | Unique elements                             | Key-value pairs                                      |      |
| **Ordered**                 | ✅ Yes                | ❌ No                                        | ✅ Yes                                                |      |
| **Mutable**                 | ❌ No                 | ✅ Yes                                       | ✅ Yes                                                |      |
| **Duplicate Elements**      | ✅ Allowed            | ❌ Not Allowed                               | ❌ Duplicate keys not allowed (values allowed)        |      |
| **Indexing**                | ✅ Yes                | ❌ No                                        | ❌ Access using keys                                  |      |
| **Access Method**           | Index                | Loop only                                   | Key                                                  |      |
| **Empty Syntax**            | `()`                 | `set()`                                     | `{}`                                                 |      |
| **Mathematical Operations** | ❌ No                 | ✅ Yes (`                                    | `, `&`, `-`, `^`)                                    | ❌ No |
| **Common Methods**          | `count()`, `index()` | `add()`, `remove()`, `discard()`, `union()` | `keys()`, `values()`, `items()`, `update()`, `pop()` |      |
| **Best Used For**          | Fixed data           | Unique values                               | Related data (key-value pairs)                       |      |
