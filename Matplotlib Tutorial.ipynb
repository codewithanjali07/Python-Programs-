# 🟢 BEGINNER LEVEL

### 1.	Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Line Plot
plt.figure(figsize=(8,5))

plt.plot(df["carat"], df["price"],
         marker="o",
         linestyle="-",
         linewidth=2)

plt.title("Line Plot of Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.grid(True)

plt.show()
```

### 2.	Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Scatter Plot
plt.figure(figsize=(8,5))

plt.scatter(df["carat"], df["price"])

plt.title("Scatter Plot of Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")

plt.grid(True)
plt.show()
```

### 3.	Bar Chart (Vertical)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Average price by cut
avg_price = df.groupby("cut")["price"].mean()

# Vertical Bar Chart
plt.figure(figsize=(8,5))

plt.bar(avg_price.index, avg_price.values)

plt.title("Average Diamond Price by Cut")
plt.xlabel("Cut")
plt.ylabel("Average Price")

plt.show()
```

### 4.	Horizontal Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Average price by color
avg_price = df.groupby("color")["price"].mean()

# Horizontal Bar Chart
plt.figure(figsize=(8,5))

plt.barh(avg_price.index, avg_price.values)

plt.title("Average Diamond Price by Color")
plt.xlabel("Average Price")
plt.ylabel("Color")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Histogram
plt.figure(figsize=(7,5))

plt.hist(df["price"], bins=10)

plt.title("Histogram of Diamond Price")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

### 6.	Pie Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Count diamonds by cut
cut_count = df["cut"].value_counts()

# Pie Chart
plt.figure(figsize=(6,6))

plt.pie(cut_count.values,
        labels=cut_count.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Diamond Distribution by Cut")

plt.show()
```

### 10.	Multiple Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Multiple Line Plot
plt.figure(figsize=(8,5))

plt.plot(df.index[:100], df["price"][:100],
         label="Price")

plt.plot(df.index[:100], df["carat"][:100],
         label="Carat")

plt.title("Multiple Line Plot of Price and Carat")
plt.xlabel("Diamond Index")
plt.ylabel("Values")
plt.legend()

plt.grid(True)
plt.show()
```

### 11.	Grouped Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = sns.load_dataset("diamonds")

# Average price by cut and color
grouped = df.groupby(["cut", "color"])["price"].mean().unstack()

x = np.arange(len(grouped.index))
width = 0.12

plt.figure(figsize=(10,5))

for i, color in enumerate(grouped.columns):
    plt.bar(x + i*width, grouped[color], width, label=color)

plt.title("Average Diamond Price by Cut and Color")
plt.xlabel("Cut")
plt.ylabel("Average Price")
plt.xticks(x + width*3, grouped.index)
plt.legend(title="Color")

plt.show()
```

### 12.	Stacked Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("diamonds")

# Create table
stacked_data = df.groupby(["cut", "color"]).size().unstack()

# Stacked Bar Chart
stacked_data.plot(
    kind="bar",
    stacked=True,
    figsize=(8,5)
)

plt.title("Stacked Bar Chart of Diamond Cut and Color")
plt.xlabel("Cut")
plt.ylabel("Count")
plt.legend(title="Color")

plt.show()
```

### 13.	Area Plot

### 14.	Error Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = sns.load_dataset("diamonds")

# Average price by cut
avg_price = df.groupby("cut")["price"].mean()

# Standard deviation
std_price = df.groupby("cut")["price"].std()

# Error Bar Plot
plt.figure(figsize=(8,5))

plt.errorbar(avg_price.index,
             avg_price.values,
             yerr=std_price.values,
             fmt="o-",
             capsize=5)

plt.title("Average Diamond Price by Cut")
plt.xlabel("Cut")
plt.ylabel("Average Price")

plt.grid(True)
plt.show()
```

### 18.	Polar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = sns.load_dataset("diamonds")

# Angle values
theta = np.linspace(0, 2*np.pi, 100)

# Radius values
r = df["price"][:100]

# Polar Plot
plt.figure(figsize=(6,6))

ax = plt.subplot(111, projection="polar")
ax.plot(theta, r)

plt.title("Polar Plot of Diamond Price")

plt.show()
```

### 19.	Contour Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = sns.load_dataset("diamonds")

# Take first 100 values
x = df["carat"][:100]
y = df["depth"][:100]

# Create grid
X, Y = np.meshgrid(
    np.linspace(x.min(), x.max(), 50),
    np.linspace(y.min(), y.max(), 50)
)

# Create Z values
Z = np.sin(X) + np.cos(Y)

# Contour Plot
plt.figure(figsize=(8,5))

plt.contour(X, Y, Z)

plt.title("Contour Plot of Diamonds")
plt.xlabel("Carat")
plt.ylabel("Depth")

plt.show()
```

## plt.subplots() 

It is used to create a figure and one or more subplots at the same time.

```python
import matplotlib.pyplot as plt

# Create one figure and one axes
fig, ax = plt.subplots()

# Data
x = [1, 2, 3, 4]
y = [10, 20, 15, 30]

# Plot graph
ax.plot(x, y)

# Add title
ax.set_title("Single Plot")

# Add axis labels
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")

# Show graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create 1 row and 2 columns
fig, ax = plt.subplots(1, 2, figsize=(8,4))

# First graph
ax[0].plot([1,2,3], [2,4,6])
ax[0].set_title("Graph 1")

# Second graph
ax[1].plot([1,2,3], [1,3,5])
ax[1].set_title("Graph 2")

# Adjust spacing
plt.tight_layout()

# Show graphs
plt.show()
```

## 3x2 Grid + Shared Axes

```python
import matplotlib.pyplot as plt

# Create 3 rows and 2 columns with shared X and Y axes
fig, ax = plt.subplots(3, 2, figsize=(8, 8), sharex=True, sharey=True)

# Data
x = [1, 2, 3, 4]

# Plot in each subplot
ax[0,0].plot(x, [1,2,3,4])
ax[0,0].set_title("Graph 1")

ax[0,1].plot(x, [2,3,4,5])
ax[0,1].set_title("Graph 2")

ax[1,0].plot(x, [3,4,5,6])
ax[1,0].set_title("Graph 3")

ax[1,1].plot(x, [4,5,6,7])
ax[1,1].set_title("Graph 4")

ax[2,0].plot(x, [5,6,7,8])
ax[2,0].set_title("Graph 5")

ax[2,1].plot(x, [6,7,8,9])
ax[2,1].set_title("Graph 6")

# Adjust spacing
plt.tight_layout()

# Display graphs
plt.show()
```

## Adjusting Spacing

```python
import matplotlib.pyplot as plt

# Create 2x2 subplots
fig, ax = plt.subplots(2, 2, figsize=(8,6))

# Graph 1
ax[0,0].plot([1,2,3], [2,4,6])
ax[0,0].set_title("Graph 1")

# Graph 2
ax[0,1].plot([1,2,3], [1,3,5])
ax[0,1].set_title("Graph 2")

# Graph 3
ax[1,0].plot([1,2,3], [6,4,2])
ax[1,0].set_title("Graph 3")

# Graph 4
ax[1,1].plot([1,2,3], [5,5,5])
ax[1,1].set_title("Graph 4")

# Adjust spacing manually
plt.subplots_adjust(wspace=0.4, hspace=0.6)

# Show graph
plt.show()
```

### Figure Facecolor vs Axes Facecolor vs Background

```python
import matplotlib.pyplot as plt

# Create figure with background color
fig = plt.figure(facecolor="lightgray")

# Create axes
ax = fig.add_subplot(111)

# Set plot area background color
ax.set_facecolor("lightyellow")

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Plot graph
ax.plot(x, y)

# Add title
ax.set_title("Figure vs Axes Facecolor")

# Add labels
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")

# Show graph
plt.show()
```

### Border / Spines Customization

```python
import matplotlib.pyplot as plt

# Create figure and axes
fig, ax = plt.subplots()

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Plot graph
ax.plot(x, y)

# Change border (spines) color
ax.spines["top"].set_color("red")
ax.spines["bottom"].set_color("blue")
ax.spines["left"].set_color("green")
ax.spines["right"].set_color("purple")

# Change border thickness
ax.spines["top"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)

# Add title
ax.set_title("Border / Spines Customization")

# Add labels
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")

# Show graph
plt.show()
```

### Tick Style

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create line plot
plt.plot(x, y)

# Customize tick style
plt.tick_params(
    axis="both",      # Apply to both axes
    direction="in",   # Tick direction
    length=8,         # Tick length
    width=2,          # Tick thickness
    colors="red",     # Tick color
    labelsize=12      # Tick label size
)

# Add title
plt.title("Tick Style Example")

# Add labels
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Show graph
plt.show()
```

### Legend Position & Multiple Legends

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

# Plot first line
plt.plot(x, y1, label="Sales")

# Plot second line
plt.plot(x, y2, label="Profit")

# Add title
plt.title("Legend Position Example")

# Display legend at upper left
plt.legend(loc="upper left")

# Show graph
plt.show()
```

### Project 1: Restaurant Bill Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Create 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(11, 8))

# Dashboard Title
fig.suptitle("Restaurant Bill Analysis Dashboard",
             fontsize=15,
             fontweight="bold")

# 1. Distribution of Total Bill (Histogram)

axes[0, 0].hist(df["total_bill"],
                bins=20,
                color="steelblue",
                edgecolor="white")

axes[0, 0].set_title("Total Bill Distribution")
axes[0, 0].set_xlabel("Total Bill ($)")
axes[0, 0].set_ylabel("Frequency")

# 2. Average Bill by Day (Bar Chart)

bill_by_day = df.groupby("day", observed=True)["total_bill"].mean().reindex(
    ["Thur", "Fri", "Sat", "Sun"]
)

axes[0, 1].bar(bill_by_day.index,
               bill_by_day.values,
               color="darkorange")

axes[0, 1].set_title("Average Bill by Day")
axes[0, 1].set_xlabel("Day")
axes[0, 1].set_ylabel("Average Bill ($)")

# 3. Bill vs Party Size (Scatter Plot)
-
axes[1, 0].scatter(df["size"],
                   df["total_bill"],
                   alpha=0.5,
                   color="seagreen")

axes[1, 0].set_title("Bill vs Party Size")
axes[1, 0].set_xlabel("Party Size")
axes[1, 0].set_ylabel("Total Bill ($)")

# 4. Tip Percentage Distribution (Histogram)

tip_percent = (df["tip"] / df["total_bill"]) * 100

axes[1, 1].hist(tip_percent,
                bins=15,
                color="purple",
                edgecolor="white")

axes[1, 1].set_title("Tip Percentage Distribution")
axes[1, 1].set_xlabel("Tip Percentage (%)")
axes[1, 1].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show dashboard
plt.show()
```

### Project 2: Customer Spending Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd

# Customer dataset
data = {
    "Customer": ["A", "B", "C", "D", "E", "F"],
    "Age": [22, 35, 28, 45, 30, 40],
    "Spending": [1500, 3200, 2100, 4500, 2800, 3900],
    "Visits": [5, 8, 6, 10, 7, 9]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create 2x2 Dashboard
fig, axes = plt.subplots(2, 2, figsize=(11, 8))

# Dashboard Title
fig.suptitle("Customer Spending Analysis Dashboard",
             fontsize=15,
             fontweight="bold")

# ----------------------------------------------------
# 1. Spending Distribution (Histogram)
# ----------------------------------------------------
axes[0, 0].hist(df["Spending"],
                bins=5,
                color="steelblue",
                edgecolor="white")

axes[0, 0].set_title("Spending Distribution")
axes[0, 0].set_xlabel("Spending (₹)")
axes[0, 0].set_ylabel("Frequency")

# ----------------------------------------------------
# 2. Spending by Customer (Bar Chart)
# ----------------------------------------------------
axes[0, 1].bar(df["Customer"],
               df["Spending"],
               color="darkorange")

axes[0, 1].set_title("Customer Spending")
axes[0, 1].set_xlabel("Customer")
axes[0, 1].set_ylabel("Spending (₹)")

# ----------------------------------------------------
# 3. Age vs Spending (Scatter Plot)
# ----------------------------------------------------
axes[1, 0].scatter(df["Age"],
                   df["Spending"],
                   color="seagreen",
                   alpha=0.7)

axes[1, 0].set_title("Age vs Spending")
axes[1, 0].set_xlabel("Age")
axes[1, 0].set_ylabel("Spending (₹)")

# ----------------------------------------------------
# 4. Customer Visits (Pie Chart)
# ----------------------------------------------------
axes[1, 1].pie(df["Visits"],
               labels=df["Customer"],
               autopct="%1.1f%%",
               startangle=90)

axes[1, 1].set_title("Customer Visits")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show dashboard
plt.show()
```

### Project 3: Tip Prediction Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")

# Create 2x2 Dashboard
fig, axes = plt.subplots(2, 2, figsize=(11, 8))

# Dashboard Title
fig.suptitle("Tip Prediction Visualization Dashboard",
             fontsize=15,
             fontweight="bold")

# ----------------------------------------------------
# 1. Total Bill Distribution (Histogram)
# ----------------------------------------------------
axes[0, 0].hist(df["total_bill"],
                bins=20,
                color="steelblue",
                edgecolor="white")

axes[0, 0].set_title("Total Bill Distribution")
axes[0, 0].set_xlabel("Total Bill ($)")
axes[0, 0].set_ylabel("Frequency")

# ----------------------------------------------------
# 2. Tip Distribution (Histogram)
# ----------------------------------------------------
axes[0, 1].hist(df["tip"],
                bins=15,
                color="orange",
                edgecolor="white")

axes[0, 1].set_title("Tip Distribution")
axes[0, 1].set_xlabel("Tip ($)")
axes[0, 1].set_ylabel("Frequency")

# ----------------------------------------------------
# 3. Total Bill vs Tip (Scatter Plot)
# ----------------------------------------------------
axes[1, 0].scatter(df["total_bill"],
                   df["tip"],
                   color="green",
                   alpha=0.6)

axes[1, 0].set_title("Total Bill vs Tip")
axes[1, 0].set_xlabel("Total Bill ($)")
axes[1, 0].set_ylabel("Tip ($)")

# ----------------------------------------------------
# 4. Average Tip by Day (Bar Chart)
# ----------------------------------------------------
tip_by_day = df.groupby("day", observed=True)["tip"].mean().reindex(
    ["Thur", "Fri", "Sat", "Sun"]
)

axes[1, 1].bar(tip_by_day.index,
               tip_by_day.values,
               color="purple")

axes[1, 1].set_title("Average Tip by Day")
axes[1, 1].set_xlabel("Day")
axes[1, 1].set_ylabel("Average Tip ($)")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show dashboard
plt.show()
```
